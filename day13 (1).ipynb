```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('taxis')
df
sns.scatterplot(x=df['fare'],y=df['total'],palette=['orange','black'],hue=df['payment'])
plt.show()
    
```
