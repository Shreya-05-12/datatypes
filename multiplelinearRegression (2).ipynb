```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
cars[['HP','WT','VOL','SP']]=sc.fit_transform(cars[['HP','WT','VOL','SP']])
```

```python
np.round(cars.describe(),2)
```

```python

import statsmodels.formula.api as smf

ml_v=smf.ols('MPG~VOL',data=cars).fit()
ml_v.rsquared
```

```python
ml_v.params
```

```python
wt_m=smf.ols('MPG~WT',data=cars).fit()
wt_m.rsquared
```

```python
wv=smf.ols('MPG~WT+VOL',data=cars).fit()
wv.rsquared
```

```python
sp_m=smf.ols('MPG~SP',data=cars).fit()
sp_m.rsquared
```

```python
hp_m=smf.ols('MPG~HP',data=cars).fit()
hp_m.rsquared
```

```python
model=smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
```

```python
model.rsquared
```

```python
model.params
```
